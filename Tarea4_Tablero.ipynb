Para ver en que punto nuestra matriz se converge, elevaremos a la potencia n.
Para el ejercicio se utlizara n=100

```python
import numpy as np
A = np.array([
    [1/4, 0,   1/4, 1/2],
    [1/2, 1/4, 0,   1/4],
    [1/4, 1/2, 1/4, 0  ],
    [0,   1/4, 1/2, 1/4]
], dtype=float)
print ('Matriz inicial:')
print(A)

def potencia_matriz(A, n): #matriz a la potencia n
    M = np.eye(A.shape[0])
    for _ in range(n):
        M = A @ M
    return M
n = 100
An = potencia_matriz(A, n)
print("-----------------------------------------------")
print(f"Matriz obtenidad con {n} iteraciones (A^{n}):\n")
print(np.round(An, 6))
```
