* <font color='salmon'> **Operaciones básicas:**

1. union(G, H[, renombrar]): Une los grafos, nodos con el mismo nombre se fusionan.
2. disjoint_union(G, H): Une los grafos siempre como disjuntos, nodos de H renumerados automáticamente.
3. cartesian_product(G, H): Devuelve el producto cartesiano de G y H.
4. create_empty_copy(G[, con_datos]): Devuelve una copia del gráfico G con todos los bordes eliminados.

```python
import networkx as nx
#Creamos dos gráficos (G y H)
G = nx.Graph()
H = nx.Graph()
G.add_edge('A','B', capacity=2)
G.add_edge('A','C', capacity=2)
G.add_edge('B','D', capacity=2)
G.add_edge('C','D', capacity=2)
G.add_edge('B','C', capacity=2)
H.add_edge('1','2', capacity=2)
H.add_edge('2','3', capacity=2)
H.add_edge('2','4', capacity=2)
```

<font color='palevioletred'> **1)**

```python
G_union=nx.union(G, H, rename=('G-', 'H-'))
plt.figure(figsize=(5,5))
nx.draw(G_union, with_labels=True, node_color="lightpink",
       node_size=1000, font_size=12, font_weight="bold",
        arrowsize=50, edgecolors="black")
plt.show()
```

<font color='palevioletred'> **2)**

```python
G_union = nx.disjoint_union(G, H)
plt.figure(figsize=(3,3))
nx.draw(G_union, with_labels=True, node_color="lightpink",
       node_size=1000, font_size=12, font_weight="bold",
        arrowsize=50, edgecolors="black")
plt.show()
```

<font color='palevioletred'> **3)**

```python
G_car=nx.cartesian_product(G, H)
plt.figure(figsize=(3,3))
nx.draw(G_car, with_labels=True, node_color="lightpink",
       node_size=1000, font_size=5, font_weight="bold",
        arrowsize=50, edgecolors="black")
plt.show()
```

<font color='palevioletred'> **4)**

```python
H = nx.create_empty_copy(G, with_data=True)
plt.figure(figsize=(5,5))
nx.draw(H, with_labels=True, node_color="lightpink",
        node_size=1000, font_size=12, font_weight="bold",
        arrowsize=14, edgecolors="black")
plt.show()
```

* <font color='salmon'> **Conectividad:**

* En grafos dirigidos tenemos dos formas:


Conectividad fuerte (strongly connected): Todos los nodos son alcanzables entre sí siguiendo las flechas.

**nx.is_strongly_connected(G)**

Conectividad débil (weakly connected): Todos los nodos están conectados si ignoramos la dirección de las flechas.

**nx.is_weakly_connected(G)**

```python
H = nx.DiGraph()
H.add_edges_from([("A", "B"), ("B", "C")])
plt.figure(figsize=(4,4))
nx.draw(H, with_labels=True, node_color="lightblue",
        node_size=2000, font_size=15, font_weight="bold",
        arrowsize=20, edgecolors="black")
plt.show()

print("¿Conexión fuerte?:", nx.is_strongly_connected(H))
print("¿Conexión débil?:", nx.is_weakly_connected(H))
```

* <font color='salmon'> **Aristas múltiples:**

Para varias aristas en la misma dirección entre dos nodos se ulitiza a nx.MultiDiGraph().

```python
G = nx.MultiDiGraph()
G.add_nodes_from([1, 2, 3, 4])

# Añadir múltiples aristas entre los mismos nodos con diferentes atributos
G.add_edge(1, 2, peso=5, capacidad=10)  #el peso es el costo/distancia por moverse de una arista a otra
G.add_edge(2, 1, peso=7, capacidad=8)   # segunda arista entre 1 y 2
G.add_edge(1, 3, peso=3, capacidad=15)
G.add_edge(2, 4, peso=2, capacidad=7)
G.add_edge(3, 4, peso=4, capacidad=12)
G.add_edge(4, 3, peso=6, capacidad=5)   # segunda arista entre 3 y 4

# Posiciones de los nodos
pos = nx.spring_layout(G)
# Dibujar nodos
nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=700)
nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold')

# Dibujar aristas
nx.draw_networkx_edges(G, pos, arrowstyle='->', arrowsize=20)

print("Atributos de las aristas:")
for u, v, key, data in G.edges(keys=True, data=True):
    print(f"Arista {u} -> {v}, key={key}, peso={data['peso']}, capacidad={data['capacidad']}")
# Mostrar grafo
plt.axis('off')
plt.show()
```

* <font color='salmon'> **Árbol de expansión mínimo**

---



**minimum_spanning_tree()** :
Esta función se utiliza en algoritmos de grafos, como los de Kruskal y Prim, para obtener este árbol de expansión con peso mínimo de un grafo ponderado y conexo.

* <font color='darkorange'>Ejemplo:

```python
import networkx as nx
import matplotlib.pyplot as plt
#EJERCICIO DE CLASE
# Creamos un grafo
G = nx.Graph()
G.add_nodes_from(["1", "2", "3", "4", "5","6"])
G.add_edge('1', '2', weight=1)
G.add_edge('1', '3', weight=5)
G.add_edge('1', '4', weight=7)
G.add_edge('2', '4', weight=4)
G.add_edge('2', '3', weight=6)
G.add_edge('2', '5', weight=3)
G.add_edge('3', '4', weight=5)
G.add_edge('3', '6', weight=10)
G.add_edge('4', '6', weight=3)
G.add_edge('4', '5', weight=8)
G.add_edge('1', '5', weight=9)
# Dibujamos el grafo
pos = nx.spring_layout(G, seed=42) #seed=42  nos permite tener siempre la misma disposición de los nodos
plt.figure(figsize=(5, 5))
nx.draw(G, pos, with_labels=True, node_color='greenyellow', node_size=600, font_weight='bold')
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.title("Grafo original", size=15)
plt.show()

```

```python
#Utilizamos la función para obtener el árbol
arbol = nx.minimum_spanning_tree(G)
# Dibujo del árbol
plt.figure(figsize=(5, 5))
nx.draw(arbol, pos, with_labels=True, node_color='lightgreen', node_size=600, font_weight='bold')
labels = nx.get_edge_attributes(arbol, 'weight')
nx.draw_networkx_edge_labels(arbol, pos, edge_labels=labels)
plt.title(f"Árbol de Expansión Mínima", size=15)
plt.show()

print("Aristas del Árbol de Expansión Mínima:")
for u, v, data in arbol.edges(data=True):
    print(f"({u}, {v}) con peso {data['weight']}")
print(f"\nPeso total del árbol: {arbol.size(weight='weight')}")
```

Se obtuvo el mismo resultado que en clase.

* <font color='salmon'> **Ruta más corta**

---



**nx.shortest_path():** Esta función puede manejar tanto grafos no ponderados (donde la ruta más corta es la que tiene menos aristas) como grafos ponderados (**donde la ruta más corta es la que tiene la menor suma de pesos en las aristas**).

**nx.shortest_path_length(G, source, target, weight=None)**: Calcula y devuelve la longitud (la suma de los pesos) de la ruta más corta.

**nx.dijkstra_path(G, source, target)** : Utiliza el algoritmo de Dijkstra específicamente para grafos ponderados.

 <font color='darkorange'> **Ejemplo:**

```python
import networkx as nx

# 1. Creamos el gráfico
G = nx.Graph()
G.add_nodes_from(["A", "B", "C", "D", "E"])
G.add_edge("A", "B", weight=4)
G.add_edge("A", "C", weight=2)
G.add_edge("B", "D", weight=5)
G.add_edge("C", "D", weight=8)
G.add_edge("C", "E", weight=10)
G.add_edge("D", "E", weight=2)
# Dibujamos el grafo
pos = nx.spring_layout(G, seed=42) #seed=42  nos permite tener siempre la misma disposición de los nodos
plt.figure(figsize=(5, 5))
nx.draw(G, pos, with_labels=True, node_color='greenyellow', node_size=600, font_weight='bold')
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.show()
```

```python
# Ruta más corta entre A y E
ruta_corta = nx.shortest_path(G, source="A", target="E", weight="weight")
print(f"La ruta más corta de A a E es: {ruta_corta}")
longitud = nx.shortest_path_length(G, source="A", target="E", weight="weight")
print(f"La longitud total de la ruta es: {longitud}")
```

 <font color='darkorange'> **Ejemplo: Algoritmo visto en clase**

```python
#EJERCICIO CLASE
#Creamos el grafo
G = nx.DiGraph()
#Añadimos nodos
G.add_edge("1", "2", weight=100)
G.add_edge("1", "3", weight=30)
G.add_edge("2", "3", weight=20)
G.add_edge("3", "4", weight=10)
G.add_edge("3", "5", weight=60)
G.add_edge("4", "2", weight=15)
G.add_edge("4", "5", weight=50)
origen = "1"
destino = "5"
ruta_dijkstra = nx.dijkstra_path(G, origen, destino, weight="weight")
print(f"La ruta más corta de {origen} a {destino} es: {ruta_dijkstra}")
longitud_dijkstra = nx.dijkstra_path_length(G, origen, destino, weight="weight")
print(f"La longitud total (suma de pesos) de la ruta es: {longitud_dijkstra}")
pos = nx.spring_layout(G)
etiquetas_aristas = nx.get_edge_attributes(G, 'weight')
#Resaltamos la ruta más corta
aristas_ruta = []

for i in range(len(ruta_dijkstra) - 1):
    arista = (ruta_dijkstra[i], ruta_dijkstra[i+1])
    aristas_ruta.append(arista)
plt.figure(figsize=(5, 5))
nx.draw(G,pos,with_labels=True, node_color='lightgreen', node_size=600, edge_color='black')
nx.draw_networkx_edges(G, pos, edgelist=aristas_ruta, edge_color='deepskyblue', width=3) #ruta
nx.draw_networkx_edge_labels(G, pos, edge_labels=etiquetas_aristas)
plt.title(f"Ruta más corta de {origen} a {destino}")
plt.show()
```

Se obtuvo el mismo resultado que en clase.

* <font color='salmon'> **Flujo máximo**

---



**nx.maximum_flow():** Esta función calcula la cantidad máxima de "flujo" que puede viajar desde un nodo de origen hasta un nodo de destino en una red con aristas de capacidad limitada.

<font color='darkorange'> **Ejemplo:**

```python
#EJERCICIO CLASE
import networkx as nx
import matplotlib.pyplot as plt
#Creamos el gráfico
G = nx.DiGraph()
G.add_edge('A', 'B', capacity=9)
G.add_edge('A', 'C', capacity=7)
G.add_edge('B', 'E', capacity=2)
G.add_edge('B', 'D', capacity=7)
G.add_edge('C', 'D', capacity=4)
G.add_edge('C', 'E', capacity=6)
G.add_edge('D', 'E', capacity=3)
G.add_edge('D', 'F', capacity=6)
G.add_edge('E', 'F', capacity=9)
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightgreen', node_size=600, font_weight='bold')
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.show()
origen = 'A'
destino = 'F'
#Flujo maximo
valor_flujo, flujo_dict = nx.maximum_flow(G, origen, destino)
print(f"El valor del flujo máximo desde {origen} a {destino} es: {valor_flujo}")
print("El flujo de cada arista es:")
print(flujo_dict)
```

Se obtuvo el mismo resultado que en clase.
