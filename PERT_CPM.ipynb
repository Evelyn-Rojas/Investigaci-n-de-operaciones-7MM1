<font color='salmon'>**Método de la ruta crítica**
> La duración del proyecto (estimada) es igual a la longitud de la ruta más larga a través de la red del proyecto. Esta trayectoria más larga se llama **ruta crítica**.

<font color='salmon'>**Trueques entre tiempo y costo de actividades individuales**
> El quiebre de una actividad se refiere a tomar medidas especiales y costosas para reducir la duración.

>El quiebre de un proyecto implica la acelaración de varias de sus actividades para reducir su duración.



Una fabrica reproductores de DVD de uso comercial está considerando la producción de reproductores de DVD para uso casero, para llevarlo a cabo se ha planteado un proyecto con 7 actividades y se desea terminar en 8 semanas al menor costo posible.

1.png

```python
import networkx as nx
import matplotlib.pyplot as plt
G = nx.Graph()
G.add_nodes_from(["Inicio","A", "B", "C", "D", "E","F","G", "Fin"])
G.add_edge('Inicio', 'A')
G.add_edge('Inicio', 'B')
G.add_edge('Inicio', 'C')
G.add_edge('A', 'D', )
G.add_edge('B', 'E', )
G.add_edge('C', 'F', )
G.add_edge('D', 'G', )
G.add_edge('E', 'G', )
G.add_edge('G', 'Fin')
G.add_edge('F', 'Fin')
pos = nx.spring_layout(G)
plt.figure(figsize=(5, 5))
nx.draw(G,pos, with_labels=True, node_color='greenyellow', node_size=600, font_weight='bold')
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.title("Red del proyecto")
```

```python
import networkx as nx
import matplotlib.pyplot as plt
G = nx.Graph()
G.add_nodes_from(["Inicio", "A", "B", "C", "D", "E", "F", "G", "Fin"])
G.add_edge('Inicio', 'A')
G.add_edge('Inicio', 'B')
G.add_edge('Inicio', 'C')
G.add_edge('A', 'D')
G.add_edge('B', 'E')
G.add_edge('C', 'F')
G.add_edge('D', 'G')
G.add_edge('E', 'G')
G.add_edge('G', 'Fin')
G.add_edge('F', 'Fin')
pos = nx.spring_layout(G, seed=42)
plt.figure(figsize=(5, 5))
nx.draw_networkx_nodes(G, pos, node_color='greenyellow', node_size=600)
ruta1 = ['A', 'D', 'G']
ruta2 = ['B', 'E', 'G']
nx.draw_networkx_nodes(G, pos, nodelist=ruta1, node_color='aquamarine', node_size=600)
nx.draw_networkx_nodes(G, pos, nodelist=ruta2, node_color='violet', node_size=600)
nx.draw_networkx_edges(G, pos, edge_color='black')
ruta11 = [('A', 'D'), ('D', 'G')]
ruta22= [('B', 'E'), ('E', 'G')]
nx.draw_networkx_edges(G, pos, edgelist=ruta11, edge_color='aquamarine', width=2)
nx.draw_networkx_edges(G, pos, edgelist=ruta22, edge_color='violet', width=2)
nx.draw_networkx_labels(G, pos, font_weight='bold')
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.title("Rutas críticas")
plt.show()
```

Aceleramos E  por dos semanas y D por otras dos semanas.

<font color='darkorange'>Costo= 150+150+100+10950=11350

El nuevo tiempo de proyecto es de 10 semanas.

Aceleramos G

<font color='darkorange'>Costo=150+150+100+250+10950=11850

**El tiempo final será de 8 semanas con un costo de 11850.**
