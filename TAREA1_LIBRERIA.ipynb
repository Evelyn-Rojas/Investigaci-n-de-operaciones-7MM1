> Un nodo Graphes una colección de nodos (vértices) junto con pares de nodos identificados (llamados aristas, enlaces, etc.).

* <font color='salmon'> **Importar libreria**


```python
import networkx as nx
import matplotlib.pyplot as plt
```

* <font color='salmon'> **Creación del gráfico no dirigdo vacío**

```python
G = nx.Graph()
```

* <font color='salmon'> **Añadir nodos**

```python
G.add_edge('A','B', capacity=2) #Recordar que capacity es cuantos arcos puede tener el nodo
G.add_edge('A','C', capacity=2)
G.add_edge('B','D', capacity=2)
G.add_edge('C','D', capacity=2)
G.add_edge('B','C', capacity=2)
plt.figure(figsize=(3,3))
nx.draw(G, with_labels=True, node_color="lightpink",
       node_size=1000, font_size=15, font_weight="bold",
        arrowsize=50, edgecolors="black")
plt.show()
```

> <font color='orange'> NOTA: El orden de los informes de adyacencia corresponde al orden de adición de las aristas.

Podemos examinar los nodos y las aristas de la red, a partir de las siguientes funciones:
1. G_nodes: devuelve los nodos del gráfico
2. G_edges: devuelve una vista de todas las aristas
3. G.adj[A]: devuelve el número de aristas de A
4. G.degree[A]: devuelve un diccionario con los nodos adyacentes a A

```python
print('Nodos:',nx.nodes(G) )
print('Arcos:',nx.edges(G) )
lista=[]
lista=list(G.neighbors('A'))
print('Aristas con el nodo A:',lista)
print('Número de aristas del nodo A:',G.degree['A'] )
```

* <font color='salmon'> **Eliminar elementos de un gráfico:**

1. G.remove_node(): remueve el nodo indicado junto con todas sus aristas
2. G.remove_nodes_from(iterable): espera una lista/iterable de nodos a eliminar
3. G.remove_edge( , ): remueve el arco indicado

```python
G.remove_node('A')
plt.figure(figsize=(3,3))
nx.draw(G, with_labels=True, node_color="lightpink",
       node_size=1000, font_size=15, font_weight="bold",
        arrowsize=50, edgecolors="black")
plt.show()
```

```python
G.remove_edge('B', 'D')
plt.figure(figsize=(3,3))
nx.draw(G, with_labels=True, node_color="lightpink",
       node_size=1000, font_size=15, font_weight="bold",
        arrowsize=50, edgecolors="black")
plt.show()
```

* <font color='salmon'> **Operaciones básicas:**

1. union(G, H[, renombrar]): Une los grafos, nodos con el mismo nombre se fusionan.
2. disjoint_union(G, H): Une los grafos siempre como disjuntos, nodos de H renumerados automáticamente.
3. cartesian_product(G, H): Devuelve el producto cartesiano de G y H.
4. create_empty_copy(G[, con_datos]): Devuelve una copia del gráfico G con todos los bordes eliminados.

```python
import networkx as nx
#Creamos dos gráficos (G y H)
G = nx.Graph()
H = nx.Graph()
G.add_edge('A','B', capacity=2)
G.add_edge('A','C', capacity=2)
G.add_edge('B','D', capacity=2)
G.add_edge('C','D', capacity=2)
G.add_edge('B','C', capacity=2)
H.add_edge('1','2', capacity=2)
H.add_edge('2','3', capacity=2)
H.add_edge('2','4', capacity=2)
```

<font color='palevioletred'> **1)**

```python
G_union=nx.union(G, H, rename=('G-', 'H-'))
plt.figure(figsize=(5,5))
nx.draw(G_union, with_labels=True, node_color="lightpink",
       node_size=1000, font_size=12, font_weight="bold",
        arrowsize=50, edgecolors="black")
plt.show()
```

<font color='palevioletred'> **2)**

```python
G_union = nx.disjoint_union(G, H)
plt.figure(figsize=(3,3))
nx.draw(G_union, with_labels=True, node_color="lightpink",
       node_size=1000, font_size=12, font_weight="bold",
        arrowsize=50, edgecolors="black")
plt.show()
```

<font color='palevioletred'> **3)**

```python
G_car=nx.cartesian_product(G, H)
plt.figure(figsize=(3,3))
nx.draw(G_car, with_labels=True, node_color="lightpink",
       node_size=1000, font_size=5, font_weight="bold",
        arrowsize=50, edgecolors="black")
plt.show()
```

<font color='palevioletred'> **4)**

```python
H = nx.create_empty_copy(G, with_data=True)
plt.figure(figsize=(5,5))
nx.draw(H, with_labels=True, node_color="lightpink",
        node_size=1000, font_size=12, font_weight="bold",
        arrowsize=14, edgecolors="black")
plt.show()
```

* <font color='salmon'> **Gráficos dirigidos**

```python
G = nx.DiGraph()

# Agregar nodos
G.add_node("A")
G.add_node("B")
G.add_node("C")

# Agregar aristas dirigidas
G.add_edge("A", "B")
G.add_edge("B", "C")
G.add_edge("C", "A")
plt.figure(figsize=(4,4))
nx.draw(G, with_labels=True, node_color="lightblue",
        node_size=2000, font_size=15, font_weight="bold",
        arrowsize=20, edgecolors="black")
plt.show()
```
